## test case 2: run time
* Complexity: $O(n)$

```python
from time import time
```

```python
signal_3 = np.repeat(signal_1, 21)
signal_4 = np.repeat(signal_2, 20)
```

```python
# plot raw signal
plot_raw_signals(signal_3, signal_4)
```

```python
# fast ddtw
start = time()
distance, path = fast_ddtw(signal_3, signal_4, K = 100)
end = time()
runtime_fastddtw = end - start
print('runtime:',  runtime_fastddtw)
print('distance:', distance)
```

```python
plot_actual_search_space(signal_3, signal_4, K = 100)
```

```python
# plot path
plot_alignment_path(path, title = 'alignment_path')
```

```python
plot_path_in_search_space(signal_3, signal_4, path, K = 100)
```
